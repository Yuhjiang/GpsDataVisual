```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from utils import *
from gpsvisual import *
```

```python
path = r'D:\Programing\Python\GpsDataVisual\trip.csv'
data = pd.read_csv(path).iloc[:, 1:]
data = data[data['日期'].isin(['2018/04/17', '2018/04/18', '2018/04/19', '2018/04/20', '2018/04/21', '2018/04/22', '2018/04/23'])]
data = data[data['用户ID'] != '866111031277674']
data.head()
```

```python
data = data[(data['出发地点经纬度'] != "('0', '0')") & (data['到达地点经纬度'] != "('0', '0')")]
```

```python
def get_gps(gps, source='iOS'):
    g = gps.split("', '")
    lat = g[0][2:]
    long = g[1][:-2]
    if source == 'Android':
        gcjlong, gcjlat = bd09_to_gcj02(float(long), float(lat))
        long = str(gcjlong)
        lat = str(gcjlat)
    new_gps = long + ',' + lat
    return new_gps
```

```python
origin = []
destination = []
num = 0
for i, row in data.iterrows():
    if num >= 10:
        break
    origin.append(get_gps(row['出发地点经纬度'], row['Source']))
    destination.append(get_gps(row['到达地点经纬度'], row['Source']))
    num += 1
```

```python
add_points(origin, destination)
```
